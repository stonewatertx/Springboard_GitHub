### 3.4 Word cloud

```python
data['parsed_list']=data.title.apply(lambda x:x.split())
```

```python
stopwords = nltk.corpus.stopwords.words('english')

import string
def process(text):
    # Check characters to see if they are in punctuation
    nopunc = set(char for char in list(text) if char not in string.punctuation)
    
    # Join the characters to form the string.
    nopunc = " ".join(nopunc)
    
    # remove any stopwords if present
    return [word for word in nopunc.lower().split() if word.lower() not in stopwords]

data['parsed_list'] = data['parsed_list'].apply(process)
```

Word cloud of positive reviews

```python
# get individual words into a list
words = []
for line in data['parsed_list'][data.label=='Positive']:
    words.extend(line)
 
## create a word frequency dictionary
wordfreq = collections.Counter(words)

# draw a Word Cloud with word frequencies
wordcloud = WordCloud(background_color='white',
                      max_words=50,
                      #stopwords=stopwords
                     ).generate_from_frequencies(wordfreq)

plt.figure(figsize=(10,9))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.title('Word Cloud of Positive Reviews',fontsize = 25, pad=15)
plt.show()
```

Word cloud of negative reviews

```python
# get individual words into a list
words = []
for line in data['parsed_list'][data.label=='Negative']:
    words.extend(line)
 
# create a word frequency dictionary
wordfreq = collections.Counter(words)

# draw a Word Cloud with word frequencies
wordcloud = WordCloud(background_color='black',
                      max_words=50,
                      #stopwords=stopwords
                     ).generate_from_frequencies(wordfreq)

plt.figure(figsize=(10,9))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.title('Word Cloud of Negative Reviews',fontsize = 25, pad=15)
plt.show()
```

### Use of stopwords

Let's see what stop_words augument does in a simple example

```python
count_vectorizer = CountVectorizer(stop_words='english')
count_title = count_vectorizer.fit_transform(['how are you?','Python is a programming language','this is not good'])
```

```python
count_title.A
```

```python
count_vectorizer.get_feature_names()
```

As shown above, stop_words filter out certain words. Like 'this', 'is', 'not', 'how', etc. Because 'not' often changes the meaning of a sentence, it probably shouldn't be filtered out for this study. We'll test the model performance with and without stop_words in the modeling session

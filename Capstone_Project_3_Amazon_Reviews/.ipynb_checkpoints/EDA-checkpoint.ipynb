## word frequency

## 1. title

### 1.a positive

```python
count_vectorizer = CountVectorizer(stop_words='english')
```

```python
count_title = count_vectorizer.fit_transform(data[data.label=='Positive'].title)
```

```python
count_title
```

```python
sum_words = count_title.sum(axis=0)
words_freq_pos = [(word, sum_words[0, idx]) for word, idx in count_vectorizer.vocabulary_.items()]
words_freq_pos =sorted(words_freq_pos, key = lambda x: x[1], reverse=True)
```

```python
words_freq_pos
```

```python
df_pos = pd.DataFrame(words_freq_pos,columns=['word','count'])
```

```python
sum_words = count_title.sum(axis=0)
words_freq_neg = [(word, sum_words[0, idx]) for word, idx in count_vectorizer.vocabulary_.items()]
words_freq_neg =sorted(words_freq_neg, key = lambda x: x[1], reverse=True)
```

## Title

```python
count_vectorizer = CountVectorizer(stop_words='english')
```

```python
X_train = data.title[:300000]
y_train = data.label[:300000]
X_test = data.title[300000:]
y_test = data.label[300000:]
```

```python
count_train = count_vectorizer.fit_transform(X_train)
count_test = count_vectorizer.transform(X_test)
```

```python
count_train
```

```python
count_test
```

```python
sum_words = count_train.sum(axis=0)
words_freq = [(word, sum_words[0, idx]) for word, idx in count_vectorizer.vocabulary_.items()]
words_freq =sorted(words_freq, key = lambda x: x[1], reverse=True)
```

```python
words_freq
```

```python
nb_classifier = MultinomialNB()
nb_classifier.fit(count_train, y_train)
```

```python
pred = nb_classifier.predict(count_test)
score = accuracy_score(y_test, pred)
```

```python
score
```

```python
neg_class_prob_sorted = nb_classifier.feature_log_prob_[0, :].argsort()[::-1]
pos_class_prob_sorted = nb_classifier.feature_log_prob_[1, :].argsort()[::-1]

print(np.take(count_vectorizer.get_feature_names(), neg_class_prob_sorted[:20]))
print(np.take(count_vectorizer.get_feature_names(), pos_class_prob_sorted[:20]))
```

### 3. tf-tdf

```python
tfidf_vectorizer = TfidfVectorizer(stop_words='english', max_df=0.7)
```

```python
tfidf_train = tfidf_vectorizer.fit_transform(X_train)
tfidf_test = tfidf_vectorizer.transform(X_test)
```

```python
nb_classifier = MultinomialNB()
nb_classifier.fit(tfidf_train, y_train) # can use count_train as well.
pred = nb_classifier.predict(tfidf_test)
score = accuracy_score(y_test, pred)
score
```

```python
sum_words = tfidf_train.sum(axis=0)
words_freq = [(word, sum_words[0, idx]) for word, idx in tfidf_vectorizer.vocabulary_.items()]
words_freq =sorted(words_freq, key = lambda x: x[1], reverse=True)
```

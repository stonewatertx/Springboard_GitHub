### 1.4 Data cleaning

Drop all 0s in electricity meter reading

```python
drop1 = elec[(elec['meter_reading']!=0)]
```

Drop all freezing meter reading

```python
drop1_sorted = drop1.sort_values(['building_id','timestamp'])
drop2 = drop1_sorted[drop1_sorted.meter_reading.diff()!=0]
```

Drop outliers from individual building level

```python
bldg_list = drop2.building_id.unique()
```

```python
drop3 = drop2.reset_index()
```

```python
for i in tqdm(bldg_list):
    data_bldg = drop3[drop3.building_id==i].copy()
    z = stats.zscore(data_bldg.meter_reading.values)
    drop3.drop(data_bldg[(z<-3) | (z>3)].index,inplace=True)
```

check again if outliers and freezing data have been removed

```python
drop3[drop3.building_id==1287].meter_reading.plot(figsize=(16,3))
```

### 1.5 Save clean data 

```python
drop3.to_csv('electricity_cleaned.csv',index=False)
```

## 3. Dataset #3 Weather data

```python
weather_raw = pd.read_csv('weather_train.csv',parse_dates=['timestamp']).set_index('timestamp')
```

```python
weather_raw.head()
```

```python
weather_raw.isnull().sum()/weather_raw.shape[0]
```

### 3.1 Data cleaning

```python
weather_raw.columns
```

Only keep air temperature and dew temperature as other features are less likely impacting building energy consumption

```python
weather_raw.drop(['cloud_coverage', 'precip_depth_1_hr', 'sea_level_pressure', 'wind_direction','wind_speed'],axis=1,inplace=True)
```

Fill missing values

```python
weather = pd.DataFrame()
for i in range(16):
    a = weather_raw[weather_raw['site_id']==i]
    a = a.interpolate()
    a = a.fillna(method='ffill')
    a = a.fillna(method='bfill')
    weather = pd.concat([weather,a])
```

```python
weather.isnull().sum()
```

```python
weather.shape
```

```python
site_list = weather.site_id.unique()
```

### 3.2 Explore the weather data pattern for different sites

```python
plt.subplots(figsize=(20,10)) 
for site in site_list:
    weather[weather.site_id==site].resample('M').air_temperature.mean().plot(label='site: '+str(site),marker='*')
plt.legend()
```

### 3.3 Save clean data 

```python
weather.to_csv('weather_cleaned.csv')
```
